# Введение

Целью данной работы является эмпирическое исследование эффективности различных алгоритмов сортировки строк. В отличие от обычных числовых данных, строки сравниваются **посимвольно**, что может существенно повлиять на производительность при больших объёмах данных. Особенно это актуально в случаях, когда строки имеют длинные общие префиксы.

В рамках исследования были реализованы и протестированы как **стандартные**, так и **специализированные** алгоритмы:

- `std::sort` (на базе Introsort)
- `MergeSort`
- `MergeSort с учётом LCP` (наибольшего общего префикса)
- `Ternary QuickSort`
- `MSD Radix Sort`
- `MSD Radix с fallback на QuickSort`

Для оценки производительности каждого алгоритма были зафиксированы:
- **время выполнения (в мс)**;
- **количество посимвольных сравнений**.

Данные собирались на 4 типах входных строк:
- `RANDOM` — полностью случайные строки;
- `REVERSED` — строки, отсортированные в обратном порядке;
- `ALMOST SORTED` — почти отсортированные строки (с небольшим шумом);
- `WITH PREFIX` — строки с одинаковыми префиксами, различия только в конце.

Массивы варьировались по размеру от **100 до 3000 строк** с шагом 100. Все результаты были сохранены в файл `results.csv` и визуализированы с помощью графиков ниже.


# Графики

## Выводы по результатам эксперимента

В ходе исследования были реализованы и протестированы шесть различных алгоритмов сортировки строк:

- `std::sort` — стандартная библиотечная реализация (на основе Introsort),
- `MergeSort`,
- `MergeSort с LCP` — с оптимизацией через наибольший общий префикс,
- `Ternary QuickSort` — тернарная быстрая сортировка по символу,
- `MSD Radix Sort` — поразрядная сортировка по старшему разряду,
- `MSD Radix + Quick` — поразрядная сортировка с переключением на быструю при малом размере подмассива.

### 🔍 Анализ времени выполнения:

1. **На случайных строках (RANDOM)**:
   - `MSD Radix + Quick` и `Ternary QuickSort` показали **наилучшее время выполнения**.
   - `MergeSort LCP` был значительно медленнее — из-за постоянных вычислений префиксов, которые не дают прироста на случайных строках.

2. **На обратно отсортированных (REVERSED)**:
   - `Ternary QuickSort` показал **всплеск времени** — из-за неудачного выбора опорных символов.
   - `MSD Radix + Quick` и `std::sort` вновь были среди самых быстрых.

3. **На почти отсортированных (ALMOST)**:
   - Все алгоритмы демонстрировали **более предсказуемое поведение**, но `MergeSort LCP` по-прежнему отстаёт по времени.
   - `MSD Radix + Quick` — остаётся самым стабильным и быстрым.

4. **На строках с общим префиксом (WITHPREFIX)**:
   - `MSD Radix` и `MSD Radix + Quick` значительно выигрывают, так как быстро обрабатывают общие начальные символы.
   - `std::sort` и `MergeSort` показали **максимальное время** — из-за избыточных посимвольных сравнений префиксов.

### Анализ количества посимвольных сравнений:

1. **На всех типах данных** `MSD Radix` и `MSD Radix + Quick` демонстрируют **наименьшее количество сравнений**.
2. `std::sort` — напротив, выполняет наибольшее число сравнений, особенно на `WITHPREFIX`.
3. `MergeSort LCP` снижает количество сравнений на массивах с совпадающими префиксами, но это не компенсирует накладные вычисления LCP по времени.

## Ответы на исследовательские вопросы из задания:

**1. Почему сложность сортировки строк отличается от обычной сортировки?**  
Потому что сравнение строк требует поэлементной проверки символов — в отличие от чисел, где одно сравнение даёт результат. Это увеличивает фактическую стоимость операции сравнения.

**2. Почему используются специализированные алгоритмы?**  
Чтобы **снизить количество посимвольных сравнений**. Алгоритмы, такие как `MSD Radix` и `Ternary QuickSort`, ориентированы на структуру строки и эффективно используют информацию о префиксах.

**3. Как поведение алгоритмов зависит от структуры данных?**  
- `MSD Radix` и `Quick+Radix` особенно эффективны на данных с общими префиксами.
- `std::sort` стабилен, но проигрывает по числу сравнений.
- `MergeSort LCP` — выигрывает по количеству сравнений, но проигрывает по времени.

**4. Какие алгоритмы наиболее эффективны?**
- Общий победитель по **времени и стабильности** — `MSD Radix + Quick`.
- Победитель по **наименьшему числу сравнений** — `MSD Radix`.
- `Ternary QuickSort` эффективен на случайных строках, но нестабилен на `REVERSED`.

---

## Общий вывод

Алгоритмы, адаптированные под особенности строк (особенно `MSD Radix`), демонстрируют **лучшую масштабируемость и производительность**. Их поведение зависит от структуры данных: наличие общих префиксов, порядок строк и шум. Использование LCP может снизить количество сравнений, но не всегда приводит к выигрышу по времени.
